```python
# Setup: Initialize the password and attempt counter
correct_password = "password123" 
attempts = 0

# Mission: Implement a secure input loop
while True:
    
    user_input = input("Enter the password to unlock the system: ")
    attempts += 1  # Increment the counter
    
    # Compare the input string to the correct password
    if user_input == correct_password:
        print(f"Access Granted! It took you {attempts} attempt(s).")
        break  
    else:
        print("Incorrect password. System remains locked. Try again.")
```
